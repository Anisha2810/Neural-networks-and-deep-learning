```python
# Import PyTorch libraries
import torch                    # Core PyTorch library
import torch.nn as nn          # For neural network layers
import torch.optim as optim    # For optimization algorithms

# -------------------------------
# STEP 1: Create Dataset (OR Gate)
# -------------------------------

# Input features (2 inputs)
X = torch.tensor([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=torch.float32)

# Expected output
Y = torch.tensor([
    [0],
    [1],
    [1],
    [1]
], dtype=torch.float32)

# -------------------------------
# STEP 2: Define SLP Model
# -------------------------------

class SLP(nn.Module):
    def __init__(self):
        super(SLP, self).__init__()

        # Single linear layer
        # Input size = 2, Output size = 1
        self.linear = nn.Linear(2, 1)

    def forward(self, x):
        # Apply linear transformation
        x = self.linear(x)

        # Apply sigmoid activation function
        # Converts output to range (0,1)
        x = torch.sigmoid(x)

        return x

# Create model object
model = SLP()

# -------------------------------
# STEP 3: Define Loss and Optimizer
# -------------------------------

# Binary Cross Entropy Loss (for binary classification)
criterion = nn.BCELoss()

# Stochastic Gradient Descent optimizer
optimizer = optim.SGD(model.parameters(), lr=0.1)

# -------------------------------
# STEP 4: Training the Model
# -------------------------------

for epoch in range(1000):

    # Forward pass: compute predicted output
    outputs = model(X)

    # Compute loss
    loss = criterion(outputs, Y)

    # Backward pass: compute gradients
    optimizer.zero_grad()   # Clear previous gradients
    loss.backward()         # Backpropagation

    # Update weights
    optimizer.step()

# -------------------------------
# STEP 5: Testing the Model
# -------------------------------

print("SLP Predictions (OR Gate):")
print(model(X).detach().round())  # Round to 0 or 1
```

```python
# Import PyTorch libraries
import torch
import torch.nn as nn
import torch.optim as optim

# -------------------------------
# STEP 1: Create Dataset (XOR Gate)
# -------------------------------

# Input data
X = torch.tensor([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=torch.float32)

# Output data (XOR logic)
Y = torch.tensor([
    [0],
    [1],
    [1],
    [0]
], dtype=torch.float32)

# -------------------------------
# STEP 2: Define MLP Model
# -------------------------------

class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()

        # First layer: Input → Hidden layer
        # 2 inputs → 4 neurons
        self.fc1 = nn.Linear(2, 4)

        # Second layer: Hidden → Output
        # 4 neurons → 1 output
        self.fc2 = nn.Linear(4, 1)

    def forward(self, x):

        # Pass input through first layer
        x = self.fc1(x)

        # Apply ReLU activation (introduces non-linearity)
        x = torch.relu(x)

        # Pass through output layer
        x = self.fc2(x)

        # Apply sigmoid activation for binary output
        x = torch.sigmoid(x)

        return x

# Create model
model = MLP()

# -------------------------------
# STEP 3: Define Loss & Optimizer
# -------------------------------

# Binary Cross Entropy Loss
criterion = nn.BCELoss()

# Adam optimizer (faster than SGD)
optimizer = optim.Adam(model.parameters(), lr=0.1)

# -------------------------------
# STEP 4: Train the Model
# -------------------------------

for epoch in range(2000):

    # Forward pass
    outputs = model(X)

    # Compute loss
    loss = criterion(outputs, Y)

    # Backward pass
    optimizer.zero_grad()   # Clear old gradients
    loss.backward()         # Compute gradients

    # Update weights
    optimizer.step()

# -------------------------------
# STEP 5: Test the Model
# -------------------------------

print("MLP Predictions (XOR Gate):")
print(model(X).detach().round())  # Convert to 0/1
```

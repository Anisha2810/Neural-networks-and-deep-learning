```python
# ---------------------------------------------
# FEED FORWARD NEURAL NETWORK USING SLP
# ---------------------------------------------

# Import required libraries
import torch
import torch.nn as nn
import torch.optim as optim

# -------------------------------
# Step 1: Prepare Dataset (OR Gate)
# -------------------------------

# Input data (features)
X = torch.tensor([[0,0], [0,1], [1,0], [1,1]], dtype=torch.float32)

# Output data (labels)
Y = torch.tensor([[0], [1], [1], [1]], dtype=torch.float32)

# -------------------------------
# Step 2: Define SLP Model
# -------------------------------

class SLP(nn.Module):
    def __init__(self):
        super(SLP, self).__init__()

        # Single layer (input → output)
        self.linear = nn.Linear(2, 1)

    def forward(self, x):
        # Forward propagation
        x = self.linear(x)        # Linear transformation
        x = torch.sigmoid(x)      # Activation function
        return x

# Create model
model = SLP()

# -------------------------------
# Step 3: Loss and Optimizer
# -------------------------------

criterion = nn.BCELoss()                     # Binary Cross Entropy Loss
optimizer = optim.SGD(model.parameters(), lr=0.1)  # Gradient Descent

# -------------------------------
# Step 4: Training
# -------------------------------

for epoch in range(1000):

    outputs = model(X)           # Forward pass
    loss = criterion(outputs, Y) # Compute loss

    optimizer.zero_grad()        # Clear gradients
    loss.backward()              # Backpropagation
    optimizer.step()             # Update weights

# -------------------------------
# Step 5: Testing
# -------------------------------

print("SLP Output (OR Gate):")
print(model(X).detach().round())
```

```python
import numpy as np

class MLP:
    def __init__(self, input_size, hidden_size, output_size, lr=0.01):
        self.lr = lr  # Learning rate

        # Initialize weights randomly
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))

        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))

    def sigmoid(self, x):
        # Activation function
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        # Derivative for backpropagation
        return x * (1 - x)

    def forward(self, X):
        # Forward pass

        # Input → Hidden
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.sigmoid(self.z1)

        # Hidden → Output
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.sigmoid(self.z2)

        return self.a2

    def backward(self, X, y, output):
        # Backpropagation

        # Error at output
        error = y - output
        d_output = error * self.sigmoid_derivative(output)

        # Error at hidden layer
        error_hidden = d_output.dot(self.W2.T)
        d_hidden = error_hidden * self.sigmoid_derivative(self.a1)

        # Update weights and biases
        self.W2 += self.a1.T.dot(d_output) * self.lr
        self.b2 += np.sum(d_output, axis=0, keepdims=True) * self.lr

        self.W1 += X.T.dot(d_hidden) * self.lr
        self.b1 += np.sum(d_hidden, axis=0, keepdims=True) * self.lr

    def train(self, X, y, epochs=10000):
        # Training loop
        for _ in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output)

    def predict(self, X):
        # Predict (round to 0 or 1)
        output = self.forward(X)
        return np.round(output)


# Example: XOR problem
if __name__ == "__main__":
    X = np.array([[0,0], [0,1], [1,0], [1,1]])
    y = np.array([[0], [1], [1], [0]])

    model = MLP(input_size=2, hidden_size=4, output_size=1, lr=0.1)
    model.train(X, y, epochs=10000)

    print("Predictions:")
    print(model.predict(X))
```
